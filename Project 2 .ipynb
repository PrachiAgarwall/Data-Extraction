```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.bsb-education.com/faculty-research/professors-directory?lang=en'
```

```python
Professor_data = requests.get(url).text
```

```python
soup = BeautifulSoup(Professor_data,"html.parser")
```

```python
list_names = []
name = soup.find_all('div', {"class" :"text-style-allcaps text-weight-bold text-size-medium"})
name
for NAME in name[:38]:
    list_names.append(NAME.text)
list_names
```

```python
list_email = []
for one_name in list_names:
    Url = 'https://www.bsb-education.com/professors-directory/' + '{}'.format(one_name).replace(" ", "-").lower().replace("'","")
    Professors_data = requests.get(Url).text
    soup = BeautifulSoup(Professors_data,"html.parser")
    name = soup.find('div', {"class" :"text-weight-semibold text-color-light-purple"})
    for EMAIL in name:
        list_email.append(EMAIL.text)
list_email
```

```python
dict= {}
for i in range(0,38):
    dict[list_names[i]] = list_email[i]
df1 = pd.DataFrame.from_dict(dict, orient= 'index')
df1
```
